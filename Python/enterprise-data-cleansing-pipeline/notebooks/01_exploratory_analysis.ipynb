# Exploratory Data Analysis: Transaction Data
## Initial assessment of raw data quality

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/input/transactions_dirty.csv')
print(f'Rows: {df.shape[0]}, Columns: {df.shape[1]}')
```

```python
missing = df.isnull().sum()
print('Missing values:\n', missing[missing > 0])
```

```python
plt.figure(figsize=(10,4))
df['amount'].hist(bins=50)
plt.title('Amount Distribution')
plt.show()
```

```python
duplicates = df.duplicated(subset=['transaction_id'], keep=False)
print(f'Duplicate transaction IDs: {duplicates.sum()}')
```

## Findings
1. Missing values in customer_id, amount
2. Negative amounts present
3. Date format inconsistency
4. Email validation needed
