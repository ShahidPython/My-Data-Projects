```python
# Cell 1: Setup
import pandas as pd, numpy as np, scipy.stats as stats, statsmodels.api as sm
df = pd.read_csv('../data/processed/movies_engineered.csv')

# Cell 2: Hypothesis 1 - Genre count vs Era
print("H1: Recent movies have more genres?")
recent = df[df['year'] >= 2015]['genre_count']
older = df[df['year'] < 2015]['genre_count']
t_stat, p_val = stats.ttest_ind(recent, older, nan_policy='omit')
print(f"T-test: t={t_stat:.3f}, p={p_val:.4f}")
print(f"Recent mean: {recent.mean():.2f}, Older mean: {older.mean():.2f}")
print("Reject H0" if p_val < 0.05 else "Fail to reject H0")

# Cell 3: Correlation Matrix
numeric_cols = ['year','genre_count','cast_size','title_length']
corr_matrix = df[numeric_cols].corr()
display(corr_matrix.style.background_gradient(cmap='coolwarm', vmin=-1, vmax=1))
```
