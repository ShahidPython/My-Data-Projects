```python
# Cell 1: Setup
import pandas as pd, numpy as np, re
df = pd.read_csv('../data/raw/movies_1990_2025.csv')

# Cell 2: Feature Creation
df['genre_count'] = df['genres'].str.split(',').str.len()
df['has_director'] = df['director'].notnull().astype(int)
df['cast_size'] = df['cast'].str.split(',').str.len().fillna(0)
df['title_length'] = df['title'].str.len()
df['is_2000s'] = ((df['year'] >= 2000) & (df['year'] < 2010)).astype(int)
df['is_2010s'] = ((df['year'] >= 2010) & (df['year'] < 2020)).astype(int)
df['release_epoch'] = pd.cut(df['year'], 
                             bins=[1990,2000,2010,2020,2025],
                             labels=['90s','00s','10s','20s'])
print(f"Created {len(df.columns)} features")
display(df[['title','year','genre_count','cast_size','release_epoch']].head())

# Cell 3: Save
df.to_csv('../data/processed/movies_engineered.csv', index=False)
print("âœ“ Engineered dataset saved")
```
